```python
df["new_status"] = df2["exit_code"]
print((df["new_status"] == 0).sum())
df.head()
```

```python
def exitc_to_words(exit_code):
    slurm_error_code = exit_code >> 8
    job_exit_code = exit_code & 0xFF

    if slurm_error_code == 0:
        if job_exit_code == 0:
            status = "Pass"
        else:
            status = "Failed"
    else:
        status = "Killed"

    return status
df["new_status"] = df["new_status"].apply(exitc_to_words)
df.head()
```

```python
df["new_status"].unique()
```

```python
pass_count = (df["new_status"] == "Pass").sum()
print(pass_count)
```

```python
killed_count = (df["new_status"] == "Killed").sum()
print(killed_count)
```

```python
failed_count = (df["new_status"] == "Failed").sum()
print(failed_count)
```
