```python
import pandas as pd
import numpy as np
from google.colab import files
```

```python
df = pd.read_csv("slurm-log.csv")
df.head(10)
```

```python
df2 = pd.read_csv("data_supercloud.csv")
df2.head(20)
```

```python
df2['submit_time'] = pd.to_datetime(df2['submit_time'], unit='s').dt.tz_localize('UTC').dt.tz_convert('US/Eastern')
```

```python
df2.head(10)
```

```python
df2["submit_time"].max()
```

```python
df2["submit_time"].min()
```

```python
start_date = pd.Timestamp('2021-06-01', tz='US/Eastern')
end_date = pd.Timestamp('2021-10-02', tz='US/Eastern')

df2 = df2[(df2['submit_time'] >= start_date) & (df2['submit_time'] <= end_date)]
```

```python
df2.head()
```

```python
df2.shape
```

```python
def datetime_to_unix(data):
    return int(data.timestamp())
```

```python
df2["submit_time"] = df2["submit_time"].apply(datetime_to_unix)
```

```python
df2.head()
```

```python
df2.to_csv("data_supercloud_updated.csv")
```
