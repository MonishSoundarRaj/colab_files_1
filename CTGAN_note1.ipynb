```python
import pandas as pd
import numpy as np
```

```python
from ctgan import CTGAN
```

```python
mira_df = mira_df.drop(columns=['Unnamed: 0'])
```

```python
mira_df["node_num"].max()
```

```python
mira_df["wait_time"] = mira_df["wait_time"].astype(float)
```

```python
mira_df["cpu_num"] = mira_df["cpu_num"].astype(int)
mira_df["node_num"] = mira_df["node_num"].astype(int)
```

```python
from ctgan import CTGAN

r_data = mira_df.head(13000)

discrete_columns = [
    'job',
    'user',
    'project',
    'state',
    'gpu_num',
    'cpu_num',
    'node_num',
    'new_status'
]

ctgan = CTGAN(epochs=10)
ctgan.fit(r_data, discrete_columns)

synthetic_data = ctgan.sample(5000)
```

```python
synthetic_data.to_csv("synthetic_data.csv", index=False)
```
